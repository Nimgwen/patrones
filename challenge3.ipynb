# Siderophores Database

This database is made up of 221 siderophores and includes the name of the compound, the associated microorganism, the type and number of chelating groups it contains (such as hydroxamate, catecholate, a-hydroxycarboxylate, carboxylate, phenolate, citrate, other* ), and the molecular representations SMILES, SMARTS and SELFIES (Siderophore_DB.csv). Descriptors: The files Siderophore_DB1, Siderophore_DB2 and Siderophore_DB3 include molecular RDKit (119), Morgan fingerprints (2049) and Mordred (1827), respectively.

# Filtrar Datos y Generar Target Categórico

```python
#Filtrar y generar un nuevo dataset donde solo hay sideroforos con un solo tipo de grupo quelante

chelant_types = ['hydroxamate', 'catecholate', 'a-hydroxycarboxylate', 'carboxylate', 'phenolate', 'citrate', 'other']

# chelant_types lista de column names de los tipos de grupos quelantes en los sideroforos

df = pd.DataFrame() # empty df para inicializar el loop


for i in chelant_types:
  chelant_types.remove(i)
  df_step = db[(db[chelant_types].apply(lambda x: x==0).all(axis=1)) & (db[i] > 0)]
  df = pd.concat([df,df_step])
  chelant_types = ['hydroxamate', 'catecholate', 'a-hydroxycarboxylate', 'carboxylate', 'phenolate', 'citrate', 'other']

# Este loop selecciona un tipo de grupo quelante en cada paso ( i in chelant_types)
# Luego filtra por conjuncion (&) de dos condiciones:
# Nota se tiene que usar "&" (bitwise boolean operator) y no "and"
# Porque se necesita bitwise operators para trabajar con series de pandas (Vectores)
# "and" solo funciona con escalares (single boolean values)

# Primera condicion: 

# usa apply para verificar que todos los grupos quelantes (menos el seleccionado para este paso)
# sea iguales a 0 
# db[chelant_types].apply(lambda x: x==0) genera una mascara de valores booleanos para cada valor
# de cada fila de las columnas de chelant_types donde es False si es diferente de 0 o True si es = 0
# luego el metodo .all() verifica por fila (axis=1) si hay al menos un False para filtrar esa fila

# Segunda condicion:

# Verifica que el tipo de grupo quelante seleccionado en ese paso del loop sea > 0 

# Luego el dataframe producto del filtro en cada paso del loop es concatenado verticalmente
# df = pd.concat([df,df_step])

display(df)
```

```python
#contar la cantidad de moleculas que SOLAMENTE son hidroxamato ("hydroxmate" >0)

def count_hydroxamate_higher_than_zero(data):
    count = len(data[(data['hydroxamate'] > 0) & 
                 (data['catecholate'] == 0) & 
                 (data['a-hydroxycarboxylate'] == 0) &
                 (data['carboxylate'] == 0) &
                 (data['phenolate'] == 0) &
                 (data['citrate'] == 0) &
                 (data['other'] == 0)])
    return count


count = count_hydroxamate_higher_than_zero(db)
print(f"The number of observations with only hydroxamate value higher than 0 is: {count}")

# de esas 90 instancias 84 son hidroxamatos, lo cual hace que las clases esten demasiado desbalanceadas

# entonces vamos a resamplear para solo clasificar entre hidroxamato y catecolato
```

# Entrenar modelo KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Como el set de datos es pequeno 184 instancias y 119 predictores
# podemos usar un ratio alto de train/test como 90/10
# para intentar maximizar el la data de entrenamiento
# para intentar obtener un modelo más potente

X = ndb.drop("chelant_type", axis=1)
y = ndb["chelant_type"]

# Split the data into a 90:10 train-test ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

k = 5
knn_model = KNeighborsClassifier(n_neighbors=k)

# Train the model
knn_model.fit(X_train, y_train)

# Test the model
y_pred = knn_model.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

# Make predictions on the test set
y_pred = knn_model.predict(X_test)

# Display the classification report and confusion matrix
print("Classification Report:")
print(classification_report(y_test, y_pred))

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Matplotlib confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix - Matplotlib')
plt.show()

# el accuracy es 1, esto puede indicar overfit
# o que las metricas no son tan fiables,
# porque decidimos darle solo 10% de las instancias al test set
# para poder usar más datos (90%) para el training set
```

```python
# Una mejor solución para el bajo numero de instancias es el cross validation
# Esto permite correr el algoritmo multiples veces (folds) utilizando fracciones
# del total de instancias como test sets y el resto como training set
# tomando como resultado las metricas promedio de todos los folds

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.neighbors import KNeighborsClassifier

X = ndb.drop("chelant_type", axis=1)
y = ndb["chelant_type"]

k = 5
knn_model = KNeighborsClassifier(n_neighbors=k)

# Perform k-fold cross-validation (e.g., 5 folds)
cv_scores = cross_val_score(knn_model, X, y, cv=5)

# Perform k-fold cross-validated predictions (e.g., 5 folds)
y_pred = cross_val_predict(knn_model, X, y, cv=5)

print("Cross-validation scores:", cv_scores)
print("Mean accuracy:", cv_scores.mean())

# Calculate the confusion matrix
cm = confusion_matrix(y, y_pred)

# classification report

print(classification_report(y,y_pred))

# Matplotlib confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix - Matplotlib (Cross-Validated)')
plt.show()


#Esto muestra métricas más realistas y el modelo parece funcionar bien
```
